```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../process_results/couples.csv', sep=';')
df['ram-g'] = df.apply(lambda x: x['ram-k'] / 1000000.0, axis=1)
```

```python
df.plot(x='vcoupled', y='ram-g')
plt.show()
```

```python
df.plot(x='vcoupled', y='procesingtime-s')
plt.show()
```

```python
corr = df.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, vmin=0.0, vmax=1.0)
plt.show()
```
