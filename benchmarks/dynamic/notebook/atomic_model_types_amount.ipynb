```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('../process_results/atomic_types.csv', sep=';')
df['memory GB'] = df.apply(lambda x: x['memory KB'] / 1000000.0, axis=1)
```

```python
axe = df.plot(x='atomic types', y='memory GB', fontsize=22)
axe.set_xlabel('atomic types', fontsize=22)
axe.legend(fontsize=22)
plt.savefig('../figures/atomic_types_memory.eps')
plt.show()
```

```python
axe = df.plot(x='atomic types', y='processing time (seconds)', fontsize=20)
axe.set_xlabel('atomic types', fontsize=20)
axe.legend(fontsize=18)
plt.savefig('../figures/atomic_types_processing_time.eps')
plt.show()
```

```python
corr = df.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, vmin=0.0, vmax=1.0)
plt.savefig('../figures/atomic_types_heatmap_correlation.eps')
plt.show()
```
