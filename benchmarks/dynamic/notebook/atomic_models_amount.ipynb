```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('../process_results/atomics.csv', sep=';')
df['memory GB'] = df.apply(lambda x: x['memory KB'] / 1000000.0, axis=1)
```

```python
df.plot(x='atomics amount', y='memory GB')
plt.savefig('../figures/atomics_amount_memory.eps')
plt.show()
```

```python
df.plot(x='atomics amount', y='processing time (seconds)')
plt.savefig('../figures/atomics_amount_processing_time.eps')
plt.show()
```

```python
corr = df.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, vmin=0.0, vmax=1.0)
plt.savefig('../figures/atomics_amount_heatmap_correlation.eps')
plt.show()
```
