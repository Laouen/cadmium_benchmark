```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./result_fold_compilation.csv', sep=';')
df['ram_gb'] = df.apply(lambda x: x['ram_kb'] / 1000000.0, axis=1)
df['commit'] = df.apply(lambda x: str(x['commit']), axis=1)

df.head()
```

```python
sns.barplot(x='commit', y='ram_gb', data=df)
plt.show()
```
